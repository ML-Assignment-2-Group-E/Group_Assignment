```python
# Create the base model with ImageNet weights
base_model = VGG16(
    weights='imagenet',
    include_top=False,           # removing og classificaitons
    input_shape=(224,224,3)      # Match image dimensiotns
)
```

```python
# Freeze base model layers
base_model.trainable = False  #  freezing the base
```

```python
# Create our model by adding classification layers
model = Sequential([
    base_model,
    Flatten(),
    Dense(256, activation='relu'),  # Smaller layer + L2 , kernel_regularizer=tf.keras.regularizers.l2(0.001)
    Dropout(0.5),
    Dense(10, activation='softmax')
])
```

```python
# custom monitoring to see what the output was as it was going
# thought about that after the last calss when the guy was talking about having to wait until the training was finished.


class TrainingMonitor(Callback):
    def __init__(self):
        super().__init__()
        self.epoch_logged = set()  # Keep track of logged epochs

    def on_epoch_end(self, epoch, logs=None):
        # Skip if we've already logged this epoch
        if epoch in self.epoch_logged:
            return
        self.epoch_logged.add(epoch)

        print("\nAvailable metrics:", logs.keys())

        print(f"\nEpoch {epoch + 1} Health Check:")
        train_acc = logs.get('accuracy')  # or might be 'acc'
        val_acc = logs.get('val_accuracy')  # or might be 'val_acc'
        gap = train_acc - val_acc if train_acc and val_acc else 0

        print(f"Training Accuracy: {train_acc:.2%}" if train_acc else "Training Accuracy: Not found")
        print(f"Validation Accuracy: {val_acc:.2%}" if val_acc else "Validation Accuracy: Not found")

        if train_acc and val_acc:
            if gap > 0.1:
                print("⚠️ Warning: Large overfitting developing")
            elif gap > 0.05:
                print("⚠️ Warning: Slight overfitting developing")
            elif gap < -0.10:
                print("⚠️ Warning: Large underfitting")
            elif gap < -0.05:
                print("⚠️ Warning: Small underfitting")
            else:
                print("✓ Model balance looks good")


# CAllbacks to prevent overfitting and monitor the progress
monitor = TrainingMonitor() # above function
early_stop_loss = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True) # if no improvement in 5 epoch, kill
early_stop_accuracy = EarlyStopping(monitor='val_accuracy', patience=5, restore_best_weights=True) # if no improvement in 5 epoch, kill
checkpoint = ModelCheckpoint('best_model.keras', monitor='val_accuracy', save_best_only=True) # saving the best model trained - uses validation accuracy to specify, could swap out to any of the other metrics
reduce_lr_loss = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=1, min_lr=1e-6) # reduces learnign rate if model not improving in three epoch
reduce_lr_overfit = ReduceLROnPlateau(monitor='val_accuracy', factor=0.2, patience=3) # reduces learnign rate if model not improving in three epoch - not being used though, actually got worse results
csv_logger = CSVLogger('training_log.csv')


# Combine callbacks   callbacks = [monitor, early_stop_loss, early_stop_accuracy, checkpoint, reduce_lr_overfit, csv_logger]
callbacks = [monitor, early_stop_loss, checkpoint, reduce_lr_overfit, csv_logger]
```

```python
import numpy as np
def evaluate_training(training_results, model, valid_batches):
    """
    Comprehensive model training evaluation
    """
    # Basic metrics
    final_acc = training_results.history['accuracy'][-1]
    final_val_acc = training_results.history['val_accuracy'][-1]
    overfitting_gap = final_acc - final_val_acc

    print("\n=== Model Training Evaluation ===")

    # Accuracy Metrics
    print("\nAccuracy Metrics:")
    print(f"Training Accuracy: {final_acc:.2%}")
    print(f"Validation Accuracy: {final_val_acc:.2%}")

    # Get validation performance
    val_loss, val_acc = model.evaluate(valid_batches, verbose=0)
    print(f"\nValidation Performance:")
    print(f"Loss: {val_loss:.4f}")
    print(f"Accuracy: {val_acc:.2%}")

    # Model Health Check
    print("\nModel Health Check:")

    # Check for overfitting
    if overfitting_gap > 0.05:
        print("⚠️ WARNING: Possible overfitting detected")
        print(f"Training-Validation accuracy gap: {overfitting_gap:.2%}")
        print("Consider: Increasing dropout or reducing model complexity")
    elif overfitting_gap < -0.05:
        print("⚠️ WARNING: Possible underfitting detected")
        print("Consider: Increasing model capacity or training longer")
    else:
        print("✓ Model appears well balanced")

    # Check learning progress
    if final_acc < 0.90:
        print("⚠️ Model accuracy below 90% - might need architecture adjustments")

    # Check for unstable training
    acc_variance = np.var(training_results.history['accuracy'])
    if acc_variance > 0.01:
        print("⚠️ Unstable training detected - consider reducing learning rate")

    return {
        'training_accuracy': final_acc,
        'validation_accuracy': final_val_acc,
        'validation_loss': val_loss,
        'overfitting_gap': overfitting_gap,
        'training_stability': acc_variance
    }

# Call it after training like this:
metrics = evaluate_training(with_callbacks, model, valid_batches)
```

```python
# Add at the start of your notebook with other imports
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

def plot_training_results(training_results):
    """
    Create and save training visualizations
    """
    # Create figure with two subplots
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

    # Plot accuracy
    ax1.plot(training_results.history['accuracy'], label='Training')
    ax1.plot(training_results.history['val_accuracy'], label='Validation')
    ax1.set_title('Model Accuracy')
    ax1.set_ylabel('Accuracy')
    ax1.set_xlabel('Epoch')
    ax1.legend()

    # Plot loss
    ax2.plot(training_results.history['loss'], label='Training')
    ax2.plot(training_results.history['val_loss'], label='Validation')
    ax2.set_title('Model Loss')
    ax2.set_ylabel('Loss')
    ax2.set_xlabel('Epoch')
    ax2.legend()

    # Adjust layout and save
    plt.tight_layout()
    timestamp = datetime.now().strftime("%Y%m%d_%H%M")
    plt.savefig(f'training_plots_{timestamp}.png')
    plt.show()

# Call this after training:
plot_training_results(with_callbacks)
#plot_training_results(train_callbacks)
```
